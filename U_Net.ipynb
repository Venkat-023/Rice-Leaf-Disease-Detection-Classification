```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import cv2

# ----------------------------
# 1. Config
# ----------------------------
IMG_SIZE = 256
BATCH_SIZE = 8
AUTOTUNE = tf.data.AUTOTUNE
EPOCHS = 70

DATA_DIR = "/content/leaf_disease_segmentation"
TRAIN_IMG_DIR = os.path.join(DATA_DIR, "aug_data/train_images")
TRAIN_MASK_DIR = os.path.join(DATA_DIR, "aug_data/train_masks")
VAL_IMG_DIR   = os.path.join(DATA_DIR, "aug_data/valid_images")
VAL_MASK_DIR  = os.path.join(DATA_DIR, "aug_data/valid_masks")

# ----------------------------
# 2. Data Loading & Augmentation
# ----------------------------
def load_image(img_path, mask_path, img_size=IMG_SIZE):
    # Load image
    img = tf.io.read_file(img_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, (img_size, img_size))
    img = tf.cast(img, tf.float32) / 255.0

    # Load mask (grayscale)
    mask = tf.io.read_file(mask_path)
    mask = tf.image.decode_jpeg(mask, channels=1)
    mask = tf.image.resize(mask, (img_size, img_size))
    mask = tf.cast(mask, tf.float32)

    # Binarize mask: any pixel > 0 becomes 1
    mask = tf.where(mask > 0, 1.0, 0.0)

    return img, mask

def augment(img, mask):
    if tf.random.uniform(()) > 0.5:
        img = tf.image.flip_left_right(img)
        mask = tf.image.flip_left_right(mask)
    if tf.random.uniform(()) > 0.5:
        img = tf.image.flip_up_down(img)
        mask = tf.image.flip_up_down(mask)
    return img, mask

def get_dataset(img_dir, mask_dir, augment_data=False):
    img_paths = sorted([os.path.join(img_dir, x) for x in os.listdir(img_dir)])
    mask_paths = sorted([os.path.join(mask_dir, x) for x in os.listdir(mask_dir)])

    dataset = tf.data.Dataset.from_tensor_slices((img_paths, mask_paths))
    dataset = dataset.map(lambda x, y: load_image(x, y), num_parallel_calls=AUTOTUNE)
    if augment_data:
        dataset = dataset.map(lambda x, y: augment(x, y), num_parallel_calls=AUTOTUNE)
    dataset = dataset.shuffle(100).batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return dataset

train_ds = get_dataset(TRAIN_IMG_DIR, TRAIN_MASK_DIR, augment_data=True)
val_ds   = get_dataset(VAL_IMG_DIR, VAL_MASK_DIR, augment_data=False)

# ----------------------------
# 3. Metrics & Loss
# ----------------------------
def dice_coefficient(y_true, y_pred, smooth=1):
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth)

def iou_metric(y_true, y_pred, smooth=1):
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    union = tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)

def dice_loss(y_true, y_pred):
    return 1 - dice_coefficient(y_true, y_pred)

def bce_dice_loss(y_true, y_pred):
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    return bce + dice_loss(y_true, y_pred)

# ----------------------------
# 4. U-Net Model
# ----------------------------
def conv_block(x, filters):
    x = layers.Conv2D(filters, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Conv2D(filters, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    return x

def encoder_block(x, filters):
    c = conv_block(x, filters)
    p = layers.MaxPooling2D((2,2))(c)
    return c, p

def decoder_block(x, skip, filters):
    us = layers.Conv2DTranspose(filters, 2, strides=2, padding="same")(x)
    concat = layers.Concatenate()([us, skip])
    c = conv_block(concat, filters)
    return c

def build_unet(input_shape=(IMG_SIZE, IMG_SIZE, 3)):
    inputs = keras.Input(input_shape)
    c1, p1 = encoder_block(inputs, 64)
    c2, p2 = encoder_block(p1, 128)
    c3, p3 = encoder_block(p2, 256)
    c4, p4 = encoder_block(p3, 512)
    bn = conv_block(p4, 1024)
    d1 = decoder_block(bn, c4, 512)
    d2 = decoder_block(d1, c3, 256)
    d3 = decoder_block(d2, c2, 128)
    d4 = decoder_block(d3, c1, 64)
    outputs = layers.Conv2D(1, 1, activation="sigmoid")(d4)
    return keras.Model(inputs, outputs, name="U-Net")

model = build_unet()
model.compile(optimizer="adam", loss=bce_dice_loss, metrics=[dice_coefficient, iou_metric])
model.summary()

# ----------------------------
# 5. Training
# ----------------------------
history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS)

# ----------------------------
# 6. Visualization
# ----------------------------
def show_predictions(dataset, num=3):
    for images, masks in dataset.take(1):
        preds = model.predict(images)
        for i in range(num):
            img = images[i].numpy()
            true_mask = masks[i].numpy().squeeze()
            pred_mask = preds[i].squeeze()

            plt.figure(figsize=(12,5))

            plt.subplot(1,3,1)
            plt.imshow(img)
            plt.title("Leaf Image")
            plt.axis('off')

            plt.subplot(1,3,2)
            plt.imshow(true_mask, cmap='gray')
            plt.title("True Mask")
            plt.axis('off')

            plt.subplot(1,3,3)
            plt.imshow(img)
            plt.imshow(pred_mask > 0.5, cmap='jet', alpha=0.5)  # binary overlay
            plt.title("Predicted Mask Overlay")
            plt.axis('off')

            plt.show()

show_predictions(val_ds)

# ----------------------------
# 7. Extract ROIs
# ----------------------------
def extract_rois_from_prediction(img_path, model, threshold=0.5):
    img = cv2.imread(img_path)
    img_resized = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    inp = np.expand_dims(img_resized / 255.0, 0)
    pred_mask = model.predict(inp)[0,:,:,0]
    binary_mask = (pred_mask > threshold).astype(np.uint8) * 255

    contours, _ = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    results = img_resized.copy()
    rois = []
    for i, cnt in enumerate(contours):
        x,y,w,h = cv2.boundingRect(cnt)
        cv2.rectangle(results, (x,y), (x+w,y+h), (0,255,0), 2)
        roi = img_resized[y:y+h, x:x+w]
        rois.append(roi)
        cv2.imwrite(f"roi_{i+1}.jpg", roi)

    plt.figure(figsize=(8,8))
    plt.imshow(cv2.cvtColor(results, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.title(f"{len(contours)} diseased regions detected")
    plt.show()

    return rois
```
